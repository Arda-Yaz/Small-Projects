```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")

train_data.head()
```

```python
train_data.head(30)
```

```python
print(train_data.isnull().sum())
```

```python
print(train_data.isnull().sum())
```

```python
train_data.drop(columns = ["Cabin"], inplace =True)
test_data.drop(columns=["Cabin"], inplace=True)
```

```python
train_data["Age"] = train_data["Age"].fillna(train_data["Age"].median())
test_data["Age"] = test_data["Age"].fillna(test_data["Age"].median())
```

```python
test_data.info()
```

```python
train_data["Embarked"] = train_data["Embarked"].fillna(train_data["Embarked"].mode()[0])
```

```python
train_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

train_data["Sex"] = train_data["Sex"].map({"male" : 0 , "female": 1})
test_data["Sex"] = test_data["Sex"].map({"male" : 0, "female" : 1})
```

```python
encoder = LabelEncoder()
train_data["Embarked"] = encoder.fit_transform(train_data["Embarked"])
test_data["Embarked"] = encoder.transform(test_data["Embarked"])
```

```python
print("Missing values in train_data:\n", train_data.isnull().sum())
print("Missing values in test_data:\n", test_data.isnull().sum())
```

```python
test_data["Fare"] = test_data["Fare"].fillna(test_data["Fare"].median())
```

```python
train_data.drop(columns=["Name", "Ticket"], inplace = True)
test_data.drop(columns=["Name", "Ticket"], inplace = True)
```

```python
features = ["Pclass", "Sex", "SibSp", "Parch", "Age","Fare"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=2)
```

```python
from sklearn.metrics import accuracy_score
model.fit(X, train_data["Survived"])
y_pred = model.predict(X)
accuracy = accuracy_score(train_data["Survived"], y_pred)
```

```python
print(accuracy)
```

```python
predictions = model.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
